```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df =  pd.read_csv('/content/data.csv')
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], bins=20, kde=True)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()

```

```python
year_bins = [1900, 1950, 1970, 1990, 2000, 2010, 2020]
year_labels = ['1900-1949', '1950-1969', '1970-1989', '1990-1999', '2000-2009', '2010-2020']

sqft_bins = [0, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000]
sqft_labels = ['0-999', '1000-1499', '1500-1999', '2000-2499', '2500-2999', '3000-3499', '3500-3999', '4000-4499', '4500-5000']

df['Year Built Binned'] = pd.cut(df['yr_built'], bins=year_bins, labels=year_labels)
df['Sqft_Above Binned'] = pd.cut(df['sqft_above'], bins=sqft_bins, labels=sqft_labels)
```

```python
# Calculate bin means for Year Built
year_binned_means = df.groupby('Year Built Binned')['price'].mean()

# Calculate bin means for Sqft_Above
sqft_binned_means = df.groupby('Sqft_Above Binned')['price'].mean()

print(year_binned_means)
print(sqft_binned_means)
```

```python
df.head()
```

```python
# 3. Identify and normalize all the attributes that has numerical values.
numerical_attributes = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition','sqft_basement','yr_renovated']]
scaler = MinMaxScaler()
normalized_attributes = scaler.fit_transform(numerical_attributes)

df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition','sqft_basement','yr_renovated']] = normalized_attributes
```

```python
#4. Visualize all the processed attributes and provide suitable inference
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Normalized Numerical Attributes")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for column in numerical_attributes.columns:
    plt.hist(normalized_attributes[:, numerical_attributes.columns.get_loc(column)], bins=30, label=column)
plt.title("Normalized Numerical Attributes Distribution")
plt.legend()
plt.show()
```
